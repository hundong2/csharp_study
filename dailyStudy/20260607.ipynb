```python
using System.Net.Sockets;
using System.IO;

string socketPath = Path.Combine(Path.GetTempPath(), "infra_ipc.sock");
if (File.Exists(socketPath)) File.Delete(socketPath);

// .NET 10: 엔드포인트 바인딩 시 커널 버퍼 직접 경로 최적화 자동 적용
using var listenSocket = new Socket(AddressFamily.Unix, SocketType.Stream, ProtocolType.Unspecified);
listenSocket.Bind(new UnixDomainSocketEndPoint(socketPath));
listenSocket.Listen(50);

Console.WriteLine($"High-Performance IPC Pipe Initialized at: {socketPath}");
```
